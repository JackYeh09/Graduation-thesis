```python
#讀取csv
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/workspace/pj/sumo-rl/outputs/n001/dqn_conn0_ep5.csv")
df
```

```python
#每5秒生成多少台車
firsthr = [6.926]*720
sechr = [9.084]*720
thirhr = [7.970]*720
fourhr = [7.863]*720
```

```python
#所有車道的車總等待時間&車流量
plt.plot(df.step, df['system_total_waiting_time'].rolling(20).mean(), color="r")
plt.ylabel("Total Waiting Time(s)", color="r")
plt.tick_params(axis='y', labelcolor="r")
plt.twinx()

plt.plot(df.step, [0]+firsthr+sechr+thirhr+fourhr, color="b")
plt.ylabel("Flows", color="b")
plt.xlabel("Time Step(s)")
plt.tick_params(axis='y', labelcolor="b")

plt.show()
```

```python
#所有車道的車平均時速&車流量
plt.plot(df.step, df['system_mean_speed'].rolling(20).mean(), color="r")
plt.ylabel("System_mean_speed(s)", color="r")
plt.tick_params(axis='y', labelcolor="r")
plt.twinx()

plt.plot(df.step, [0]+firsthr+sechr+thirhr+fourhr, color="b")
plt.ylabel("Flows", color="b")
plt.xlabel("Time Step(s)")
plt.tick_params(axis='y', labelcolor="b")
```

```python
#所有車道的擠了多少台車&車流量
plt.plot(df.step, df['system_total_queue'].rolling(20).mean(), color="r")
plt.ylabel("System_total_queue(s)", color="r")
plt.tick_params(axis='y', labelcolor="r")
plt.twinx()

plt.plot(df.step, [0]+firsthr+sechr+thirhr+fourhr, color="b")
plt.ylabel("Flows", color="b")
plt.xlabel("Time Step(s)")
plt.tick_params(axis='y', labelcolor="b")
```
